# **Formatting**

```python
# Modify Amount column 
# Remove currency sign, remove thousand separator, replace () with - sign, change to float type

pcard['Amount'] = pcard['Amount'].astype(str)
pcard['Amount'] = pcard['Amount'].str.replace('$', '')
pcard['Amount'] = pcard['Amount'].str.replace(',', '')

pcard['Amount'] = pcard['Amount'].str.replace('(', '-')
pcard['Amount'] = pcard['Amount'].str.replace(')', '')

pcard['Amount'] = pcard.Amount.astype(float)
```

```python
# Remove special characters from field names

pcard.columns = pcard.columns.str.replace(r'[^\w]', '', regex=True)
```

```python
# Change TransactionDate & PostedDate Datatype to datetime

pcard['TransactionDate'] = pd.to_datetime(pcard['TransactionDate'], format='%m/%d/%Y %H:%M')
pcard['PostedDate'] = pd.to_datetime(pcard['PostedDate'], format='%m/%d/%Y %H:%M')
```

```python
# Remove all leading & trailing blanks

for col in pcard.select_dtypes(include='object'):
    pcard[col] = pcard[col].str.strip()    
```

```python
# UNNECESSARY: capitalize all strings

for col in pcard.select_dtypes(include='object'):
    pcard[col] = pcard[col].str.upper()
```

### 1 - User shall not spend more than $5,000 per transaction

```python
pcard_1 = pcard_2014_osu[pcard_2014_osu['Amount'] > 5000].sort_values(by = 'Amount', ascending = False)
pcard_1 = pcard_1.reset_index(drop=True)

pcard_1
```

### 2 - User shall not spend more than $50,000 per year

```python
pcard_by_cardholder = pcard_2014_osu.groupby(['CardholderLastName', 'CardholderFirstInitial'])['Amount'].sum() 
pcard_by_cardholder = pcard_by_cardholder.to_frame().reset_index() 

pcard_2 = pcard_by_cardholder[pcard_by_cardholder['Amount'] > 50000].sort_values(by = 'Amount', ascending = False)
pcard_2 = pcard_2.reset_index(drop=True)

pcard_2
```

```python
print(pcard_2.shape)
print(pcard_2.dtypes)
```

### 3 - User shall not spend more than $10,000 per month without approval

```python
# NOTE
# does month datatype needs to be datetime?
```

```python
groupby_month = pcard_2014_osu.TransactionDate.dt.month

by_month = pcard_2014_osu.groupby(['CardholderLastName', 'CardholderFirstInitial', groupby_month])['Amount'].sum() 
by_month = by_month.to_frame().reset_index() 
by_month.rename(columns = {'TransactionDate':'Month'}, inplace = True)

pcard_3 = by_month[by_month['Amount'] > 10000].sort_values(by = ['Month', 'Amount'], ascending = False)
pcard_3 = pcard_3.reset_index(drop=True)

pcard_3
```

```python
print(pcard_3.shape)
print(pcard_3.dtypes)
```

### 4 - An amount more than $5,000 should not be split between two or more swipes of the card by the same person

```python
by_cardholder_by_day = pcard_2014_osu.groupby(['CardholderLastName', 'CardholderFirstInitial', 'TransactionDate', 'Vendor'])['Amount'].sum() 
by_cardholder_by_day = by_cardholder_by_day.to_frame().reset_index()

pcard_4 = by_cardholder_by_day[by_cardholder_by_day['Amount'] > 5000].sort_values(by = 'TransactionDate', ascending = True)
pcard_4 = pcard_4.reset_index(drop=True)

pcard_4
```

```python
print(pcard_4.shape)
print(pcard_4.dtypes)
```

### 5 - Continued from 4 - Count number of violations per individual

```python
count_violation = pcard_4.groupby(['CardholderLastName', 'CardholderFirstInitial']).count()
count_violation = count_violation.reset_index()
count_violation['ViolationCount'] = count_violation['Amount']

pcard_5 = count_violation.drop(columns = ['TransactionDate', 'Vendor', 'Amount'])

pcard_5 = pcard_5.sort_values(by = 'ViolationCount', ascending = False)
pcard_5 = pcard_5.reset_index(drop=True)

pcard_5
```

```python
print(pcard_5.shape)
print(pcard_5.dtypes)
```

### 6 - Bonus - Purchases should not be split between two or more cardholders

```python
by_vendor = pcard_2014_osu.groupby(['Vendor', 'TransactionDate'])['Amount'].sum() 
by_vendor = by_vendor.to_frame().reset_index()

count_customer = pcard_2014_osu.groupby(['Vendor', 'TransactionDate'])[['CardholderLastName', 'CardholderFirstInitial']].nunique()
count_customer = count_customer.reset_index()

by_vendor['UniqueCustomerCount'] = count_customer['CardholderLastName']

over_5k = by_vendor['Amount'] > 5000
multiple_customers = by_vendor['UniqueCustomerCount'] > 1

pcard_6 = by_vendor[over_5k & multiple_customers].sort_values(by = 'TransactionDate', ascending = True)
pcard_6 = pcard_6.reset_index(drop=True)
pcard_6
```

```python
print(pcard_6.shape)
print(pcard_6.dtypes)
```

### 7 - Transactions are prohibited for expenses for food and mileage while traveling. A per diem for food expenses and mileage may be claimed using a travel voucher

```python
mcc_accommodation = pcard_2014_osu[pcard_2014_osu['MerchantCategoryCodeMCC'].str.contains('HOTEL|MOTEL|RESORT|INN')]
mcc_food = pcard_2014_osu[pcard_2014_osu['MerchantCategoryCodeMCC'].str.contains('FOOD|RESTAURANT')]

merged = pd.merge(
    mcc_accommodation,
    mcc_food,
    on=['AgencyNumber', 'AgencyName', 'CardholderLastName', 'CardholderFirstInitial', 'TransactionDate'],
    suffixes=('_accommodation', '_food'))

pcard_7 = merged.sort_values(by=['CardholderLastName', 'Amount_accommodation'])
pcard_7
```
